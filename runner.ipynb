# Trainer


```python
# Runner and fitness
def run(agents,size=15,steps=400,seed=1):
    env = make("halite", configuration={"size": size, "startingHalite": ((24000/441)*size**2)//1,"episodeSteps": steps,'randomSeed':1}, debug=False)
    env.reset(len(agents))
    env.run(agents)
    return env

def fitness_halite(env):
    state = env.state[0]
    reward = state.reward
    return reward
```

```python
# Agent evaluation techniques

def double_fitness(agent, n):
    # Run 5 1v1 against comp1.py and take average fitness_halite
    res = 0
    for randomSeed in range(50,n+50):
        res += fitness_halite(run([agent,'comp1.py'],seed=random)) / n
    return res
```
